```python
# Import necessary libraries
import pandas as pd

# Load the dataset
file_path = 'instagram_data.xlsx'  # Make sure the dataset is in the same directory
data = pd.read_excel(file_path)

# Display dataset information
print("Dataset Info:")
print(data.info())  # Check column data types and non-null counts

# Display the first few rows
print("\nFirst 5 Rows of the Dataset:")
print(data.head())
```

```python
#Step 4: Handle Missing Values
#Objective:
#Decide how to deal with missing values in TEXT, list_of_tags, and list_of_mentions.
#Justify the chosen methods for handling missing data.

# Replace missing values in the TEXT column with "No Caption"
data['TEXT'] = data['TEXT'].fillna("No Caption")

# Replace missing values in list_of_tags and list_of_mentions with an empty string
data['list_of_tags'] = data['list_of_tags'].fillna("")
data['list_of_mentions'] = data['list_of_mentions'].fillna("")

# Verify that missing values have been handled
print("Missing Values After Handling:")
print(data.isnull().sum())
```

```python
# Step 5: Analyze and Visualize Data Distributions
# Objective: To explore the distributions of key numerical variables like LIKES and COMMENTS to understand their spread and detect potential outliers.

# Step 5: Visualize distributions of numerical columns

# Import seaborn and matplotlib for visualizations
import seaborn as sns
import matplotlib.pyplot as plt

# Plot the distribution of 'LIKES'
plt.figure(figsize=(12, 6))
sns.histplot(data['LIKES'], bins=50, kde=True, color='blue')
plt.title('Distribution of Likes')
plt.xlabel('Likes')
plt.ylabel('Frequency')
plt.show()

# Plot the distribution of 'COMMENTS'
plt.figure(figsize=(12, 6))
sns.histplot(data['COMMENTS'], bins=50, kde=True, color='green')
plt.title('Distribution of Comments')
plt.xlabel('Comments')
plt.ylabel('Frequency')
plt.show()
```

```python
# Step 6: Outlier Detection and Analysis
# Objective: To identify and visualize outliers in LIKES and COMMENTS to decide how to handle them.

# Import seaborn and matplotlib for visualizations
import seaborn as sns
import matplotlib.pyplot as plt

# Create boxplots for 'LIKES' to identify outliers
plt.figure(figsize=(10, 5))
sns.boxplot(x=data['LIKES'], color='blue')
plt.title('Boxplot of Likes')
plt.xlabel('Likes')
plt.show()

# Create boxplots for 'COMMENTS' to identify outliers
plt.figure(figsize=(10, 5))
sns.boxplot(x=data['COMMENTS'], color='green')
plt.title('Boxplot of Comments')
plt.xlabel('Comments')
plt.show()

# Quantify outliers using the IQR method
Q1_likes = data['LIKES'].quantile(0.25)
Q3_likes = data['LIKES'].quantile(0.75)
IQR_likes = Q3_likes - Q1_likes
lower_bound_likes = Q1_likes - 1.5 * IQR_likes
upper_bound_likes = Q3_likes + 1.5 * IQR_likes

Q1_comments = data['COMMENTS'].quantile(0.25)
Q3_comments = data['COMMENTS'].quantile(0.75)
IQR_comments = Q3_comments - Q1_comments
lower_bound_comments = Q1_comments - 1.5 * IQR_comments
upper_bound_comments = Q3_comments + 1.5 * IQR_comments

# Count outliers for likes and comments
outliers_likes = data[(data['LIKES'] < lower_bound_likes) | (data['LIKES'] > upper_bound_likes)]
outliers_comments = data[(data['COMMENTS'] < lower_bound_comments) | (data['COMMENTS'] > upper_bound_comments)]

print(f"Number of outliers in Likes: {len(outliers_likes)}")
print(f"Number of outliers in Comments: {len(outliers_comments)}")
```

```python
# Step 7: Handle Outliers
# Objective:
# To Address outliers in LIKES and COMMENTS based on their impact on the analysis.
# Choose an appropriate method:
# Cap the Outliers: Replace values above the upper bound with the upper bound.
# Remove the Outliers: Exclude rows with outliers if they are deemed non-representative.
# Step 7: Cap Outliers in LIKES and COMMENTS

# Define the bounds for LIKES
Q1_likes = data['LIKES'].quantile(0.25)
Q3_likes = data['LIKES'].quantile(0.75)
IQR_likes = Q3_likes - Q1_likes
lower_bound_likes = Q1_likes - 1.5 * IQR_likes
upper_bound_likes = Q3_likes + 1.5 * IQR_likes

# Define the bounds for COMMENTS
Q1_comments = data['COMMENTS'].quantile(0.25)
Q3_comments = data['COMMENTS'].quantile(0.75)
IQR_comments = Q3_comments - Q1_comments
lower_bound_comments = Q1_comments - 1.5 * IQR_comments
upper_bound_comments = Q3_comments + 1.5 * IQR_comments

# Cap outliers in LIKES
data['LIKES'] = data['LIKES'].clip(lower=lower_bound_likes, upper=upper_bound_likes)

# Cap outliers in COMMENTS
data['COMMENTS'] = data['COMMENTS'].clip(lower=lower_bound_comments, upper=upper_bound_comments)

# Verify the changes with boxplots
plt.figure(figsize=(10, 5))
sns.boxplot(x=data['LIKES'], color='blue')
plt.title('Boxplot of Likes After Capping Outliers')
plt.xlabel('Likes')
plt.show()

plt.figure(figsize=(10, 5))
sns.boxplot(x=data['COMMENTS'], color='green')
plt.title('Boxplot of Comments After Capping Outliers')
plt.xlabel('Comments')
plt.show()
```

```python
# Phase 2: Data Preparation
# Step 1: Feature Engineering
# Step 1.1: Extract Day of Week, Month, and Hour from the DATE column
# Objective: To convert the DATE column into new categorical features to capture temporal patterns.
# Plan:
# Extract the Day of Week (e.g., Monday, Tuesday) from the DATE column.
# Extract the Month (e.g., January, February).
# Extract the Hour (0-23) for time-based insights.
# Step 1.1: Extracting Day of Week, Month, and Hour from the DATE column

# Convert the DATE column to datetime format if not already done
data['DATE'] = pd.to_datetime(data['DATE'])

# Extract Day of Week
data['Day_of_Week'] = data['DATE'].dt.day_name()  # e.g., 'Monday', 'Tuesday'

# Extract Month
data['Month'] = data['DATE'].dt.month_name()  # e.g., 'January', 'February'

# Extract Hour
data['Hour'] = data['DATE'].dt.hour  # 0-23 format

# Display the updated dataset with new columns
print(data[['DATE', 'Day_of_Week', 'Month', 'Hour']].head())
```

```python
# Step 1.3: Create a Binary Variable for HAS_TAGS
# Objective: to identify whether a post has hashtags by checking if the list_of_tags column contains any data.
# Plan:
# Check the list_of_tags column for content.
# Create a new column HAS_TAGS where:
# 1 indicates the post has hashtags.
# 0 indicates no hashtags.

# Step 1.3: Creating a binary variable for HAS_TAGS

# Create HAS_TAGS column: 1 if list_of_tags is non-empty, otherwise 0
data['HAS_TAGS'] = data['list_of_tags'].apply(lambda x: 1 if len(str(x).strip()) > 0 else 0)

# Verify the results by displaying the first few rows
print(data[['list_of_tags', 'HAS_TAGS']].head())
```

```python
# Step 3: Data Transformation where we handle any skewed distributions or normalize data for better model performance.
# Step 3.1: Check for Skewness in LIKES and COMMENTS
# Plan: 
# Assess the skewness of LIKES and COMMENTS to decide if log transformations are necessary.
# If skewness is high, apply a log transformation to reduce the impact of extreme values.

# Step 3.1: Check skewness of LIKES and COMMENTS
from scipy.stats import skew

# Calculate skewness
likes_skewness = skew(data['LIKES'])
comments_skewness = skew(data['COMMENTS'])

# Print skewness values
print("Skewness of LIKES:", likes_skewness)
print("Skewness of COMMENTS:", comments_skewness)

# If skewness > 1 or < -1, consider log transformation
```

```python
#Steps 3.2 Based on Skewness Since the skewness for both LIKES and COMMENTS exceeds 1 (positive skew),
# applying a log transformation is necessary to reduce skewness and normalize the data distribution.
# Step 3.2: Apply log transformations to reduce skewness
# Create new columns LOG_LIKES and LOG_COMMENTS to store the transformed values.
# Use the natural logarithm for the transformation, but ensure that zeros are handled appropriately (log(0) is undefined).

# Import the numpy library
import numpy as np

# Apply log transformation to LIKES and COMMENTS, adding 1 to avoid log(0)
data['LOG_LIKES'] = (data['LIKES'] + 1).apply(np.log)
data['LOG_COMMENTS'] = (data['COMMENTS'] + 1).apply(np.log)

# Display the first few rows of the transformed columns
print(data[['LIKES', 'LOG_LIKES', 'COMMENTS', 'LOG_COMMENTS']].head())
```

```python
# Question 2
#Step 1: Create New Variables
#1. Sentiment Score
#We will use a sentiment analysis library like TextBlob to calculate the sentiment score for each post.
#2. Text Complexity Index
#We will calculate the average word length for each post as a proxy for text complexity.
#3. Engagement Ratio
#We will calculate the ratio of FOLLOWERS to FOLLOWING.

from textblob import TextBlob
import numpy as np

# 1. Sentiment Score
def calculate_sentiment(text):
    if isinstance(text, str):
        return TextBlob(text).sentiment.polarity
    return 0  # Default sentiment for empty or invalid text

data['Sentiment_Score'] = data['TEXT'].apply(calculate_sentiment)

# 2. Text Complexity Index (Average Word Length)
def calculate_avg_word_length(text):
    if isinstance(text, str):
        words = text.split()
        if len(words) > 0:
            return np.mean([len(word) for word in words])
    return 0  # Default complexity for empty or invalid text

data['Text_Complexity'] = data['TEXT'].apply(calculate_avg_word_length)

# 3. Engagement Ratio
# Avoid division by zero
data['Engagement_Ratio'] = data['FOLLOWERS'] / data['FOLLOWING'].replace(0, np.nan)
data['Engagement_Ratio'].fillna(0, inplace=True)  # Replace NaN with 0 for cases where FOLLOWING is 0

# Display sample rows to verify
print("Sample data with new variables:")
print(data[['TEXT', 'Sentiment_Score', 'Text_Complexity', 'Engagement_Ratio']].head())
```

```python
# Step 2: Regression Analysis with New Variables

# What This Does:
# Multicollinearity Check:
# Calculates the Variance Inflation Factor (VIF) for all predictors to identify multicollinearity.
# Variables with a VIF > 5 may need to be removed.
# Regression Analysis:
# Runs separate regression models for LOG_LIKES and LOG_COMMENTS, including the newly created variables.
# Model Performance:
# Evaluates the adjusted R-squared values to measure model improvement.

import numpy as np
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm

# Step 1: Initial VIF Calculation (No Variables Dropped Yet)
# Prepare feature set for initial VIF calculation
X_initial = pd.get_dummies(data, drop_first=True)  # Convert categorical variables to numeric dummies if any

# Ensure all columns are numeric
X_initial_numeric = X_initial.select_dtypes(include=[np.number])

# Add a constant for VIF calculation
X_initial_numeric = sm.add_constant(X_initial_numeric)

# Calculate VIF for each feature
vif_data_initial = pd.DataFrame()
vif_data_initial['Feature'] = X_initial_numeric.columns
vif_data_initial['VIF'] = [variance_inflation_factor(X_initial_numeric.values, i) for i in range(X_initial_numeric.shape[1])]

print("Initial Variance Inflation Factor (VIF):")
print(vif_data_initial)
```

```python
# Step 2: Drop Problematic Variables and Recalculate VIF
# We'll address the multicollinearity by:
# Dropping variables with extremely high or infinite VIF values (e.g., Afternoon, Evening, Night, Time_Check).
# Recalculating VIF to ensure the issue is resolved.

# Step 2: Drop Variables with High Multicollinearity
# Dropping 'Afternoon', 'Evening', 'Night', 'Time_Check', 'LateNight', and 'Morning' due to collinearity issues
X_reduced = X_initial.drop(columns=['Afternoon', 'Evening', 'Night', 'Time_Check', 'LateNight', 'Morning'], errors='ignore')  # Drop problematic variables

# Ensure all columns are numeric
X_reduced_numeric = X_reduced.select_dtypes(include=[np.number])

# Add a constant for VIF calculation
X_reduced_numeric = sm.add_constant(X_reduced_numeric)

# Recalculate VIF
vif_data_reduced = pd.DataFrame()
vif_data_reduced['Feature'] = X_reduced_numeric.columns
vif_data_reduced['VIF'] = [variance_inflation_factor(X_reduced_numeric.values, i) for i in range(X_reduced_numeric.shape[1])]

print("Reduced Variance Inflation Factor (VIF) After Dropping High-Collinearity Variables:")
print(vif_data_reduced)
```

```python
# Step 3: Drop Additional Variables with High Multicollinearity
# Using the reduced dataset from Step 2 as the starting point

# Drop problematic columns identified in Step 2 and Step 3
X_reduced = X_reduced.drop(columns=['Hour', 'HAS_TAGS', 'LOG_COMMENTS', 'LOG_LIKES',
                                    'Morning', 'Afternoon', 'Evening', 'Night',
                                    'Time_Check', 'LateNight'], errors='ignore')

# Ensure all columns are numeric
X_reduced_numeric = X_reduced.select_dtypes(include=[np.number])

# Add a constant for VIF calculation
X_reduced_numeric = sm.add_constant(X_reduced_numeric)

# Recalculate VIF
vif_data_reduced = pd.DataFrame()
vif_data_reduced['Feature'] = X_reduced_numeric.columns
vif_data_reduced['VIF'] = [variance_inflation_factor(X_reduced_numeric.values, i) for i in range(X_reduced_numeric.shape[1])]

print("Final Variance Inflation Factor (VIF) After Dropping All Problematic Variables:")
print(vif_data_reduced)
```

```python
# Recreate additional missing variables for both datasets

from textblob import TextBlob
import numpy as np

# Recreate missing variables for both datasets
def recreate_missing_variables(df):
    # Ensure 'TEXT' column is clean
    df['TEXT'] = df['TEXT'].fillna("").astype(str)
    
    # Sentiment_Score
    df['Sentiment_Score'] = df['TEXT'].apply(lambda x: TextBlob(x).sentiment.polarity)
    
    # Text_Complexity
    df['Text_Complexity'] = df['TEXT'].apply(lambda x: len(x.split()) / len(x) if len(x) > 0 else 0)
    
    # Engagement_Ratio
    df['Engagement_Ratio'] = (df['LIKES'] + df['COMMENTS']) / df['FOLLOWERS']
    return df

# Apply the recreation to both micro and macro datasets
micro_influencers_cleaned = recreate_missing_variables(micro_influencers_cleaned)
macro_influencers_cleaned = recreate_missing_variables(macro_influencers_cleaned)

# Encode categorical variables
columns_to_encode = ['Day of Posting', 'Timing', 'TYPE(1 PHOTO,2 VIDEO)']
micro_influencers_encoded = pd.get_dummies(
    micro_influencers_cleaned,
    columns=columns_to_encode,
    drop_first=True
)
macro_influencers_encoded = pd.get_dummies(
    macro_influencers_cleaned,
    columns=columns_to_encode,
    drop_first=True
)

# Check the cleaned datasets
print("Cleaned Micro Influencers Data:")
print(micro_influencers_cleaned.head())
print("\nCleaned Macro Influencers Data:")
print(macro_influencers_cleaned.head())
```

```python
# Check Multicollinearity: Step 3
# Multicollinearity Analysis
import statsmodels.api as sm

# Multicollinearity Analysis
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calculate_vif(df):
    # Select numeric columns only
    numeric_df = df.select_dtypes(include=[np.number])
    
    # Add constant column for VIF calculation
    numeric_df = sm.add_constant(numeric_df)
    
    vif_data = pd.DataFrame()
    vif_data['Feature'] = numeric_df.columns
    vif_data['VIF'] = [variance_inflation_factor(numeric_df.values, i) for i in range(numeric_df.shape[1])]
    
    return vif_data

# Perform VIF analysis for micro influencers
to_check_vif_micro = micro_influencers_encoded.drop(columns=['LIKES', 'COMMENTS'], errors='ignore')
vif_micro = calculate_vif(to_check_vif_micro)
print("VIF for Micro Influencers:")
print(vif_micro)

# Perform VIF analysis for macro influencers
to_check_vif_macro = macro_influencers_encoded.drop(columns=['LIKES', 'COMMENTS'], errors='ignore')
vif_macro = calculate_vif(to_check_vif_macro)
print("VIF for Macro Influencers:")
print(vif_macro)
```

```python
# Next Steps
# Visualize Regression Coefficients: Plot the coefficients for the significant predictors from each model.
# Prediction vs Actuals Visualization: Create scatter plots to show the relationship between actual and predicted values for LIKES and COMMENTS.
# Residual Analysis: Plot residuals to check for patterns that may suggest model improvements.

# Visualize and Compare Results for Micro and Macro Influencers

# Visualize and Compare Results for Micro and Macro Influencers

import statsmodels.api as sm
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

def plot_coefficients(model, title):
    coefficients = model.params[1:]  # Exclude the constant
    errors = model.bse[1:]  # Standard errors for coefficients

    plt.figure(figsize=(10, 6))
    plt.bar(coefficients.index, coefficients.values, yerr=errors, capsize=5)
    plt.xticks(rotation=45, ha='right')
    plt.title(title)
    plt.ylabel('Coefficient Value')
    plt.tight_layout()
    plt.show()

def plot_predictions_vs_actuals(model, df, target_variable, predictors, title):
    X = df[predictors]
    X = sm.add_constant(X)
    predictions = model.predict(X)

    plt.figure(figsize=(8, 8))
    plt.scatter(df[target_variable], predictions, alpha=0.6)
    plt.plot([df[target_variable].min(), df[target_variable].max()], 
             [df[target_variable].min(), df[target_variable].max()], 'r--')
    plt.title(title)
    plt.xlabel('Actual Values')
    plt.ylabel('Predicted Values')
    plt.tight_layout()
    plt.show()

def plot_residuals(model, df, target_variable, predictors, title):
    X = df[predictors]
    X = sm.add_constant(X)
    residuals = df[target_variable] - model.predict(X)

    plt.figure(figsize=(8, 6))
    plt.scatter(model.predict(X), residuals, alpha=0.6)
    plt.axhline(y=0, color='r', linestyle='--')
    plt.title(title)
    plt.xlabel('Fitted Values')
    plt.ylabel('Residuals')
    plt.tight_layout()
    plt.show()

def create_comparison_table(models, labels):
    comparison = []
    for model, label in zip(models, labels):
        coef_summary = pd.DataFrame({
            'Variable': model.params.index,
            'Coefficient': model.params.values,
            'P-Value': model.pvalues.values,
            'Standard Error': model.bse.values
        })
        coef_summary['Model'] = label
        comparison.append(coef_summary)

    comparison_df = pd.concat(comparison)
    return comparison_df

# Define predictors for regression
predictors = [
    'FOLLOWING', 'number_of_tags', 'number_of_mentions', 'Engagement_Ratio',
    'Text_Complexity', 'Sentiment_Score', 'Timing_Afternoon', 'Timing_Evening',
    'Day of Posting_Wednesday'
]

# Visualize Results
# Plot coefficients for Micro Influencers
plot_coefficients(micro_likes_model, "Micro Influencers - Likes Coefficients")
plot_coefficients(micro_comments_model, "Micro Influencers - Comments Coefficients")

# Plot coefficients for Macro Influencers
plot_coefficients(macro_likes_model, "Macro Influencers - Likes Coefficients")
plot_coefficients(macro_comments_model, "Macro Influencers - Comments Coefficients")

# Plot predictions vs actuals for Micro Influencers
plot_predictions_vs_actuals(micro_likes_model, micro_influencers_encoded, 'LIKES', predictors, "Micro Influencers - Predicted vs Actual Likes")
plot_predictions_vs_actuals(micro_comments_model, micro_influencers_encoded, 'COMMENTS', predictors, "Micro Influencers - Predicted vs Actual Comments")

# Plot predictions vs actuals for Macro Influencers
plot_predictions_vs_actuals(macro_likes_model, macro_influencers_encoded, 'LIKES', predictors, "Macro Influencers - Predicted vs Actual Likes")
plot_predictions_vs_actuals(macro_comments_model, macro_influencers_encoded, 'COMMENTS', predictors, "Macro Influencers - Predicted vs Actual Comments")

# Plot residuals for Micro Influencers
plot_residuals(micro_likes_model, micro_influencers_encoded, 'LIKES', predictors, "Micro Influencers - Residuals for Likes")
plot_residuals(micro_comments_model, micro_influencers_encoded, 'COMMENTS', predictors, "Micro Influencers - Residuals for Comments")

# Plot residuals for Macro Influencers
plot_residuals(macro_likes_model, macro_influencers_encoded, 'LIKES', predictors, "Macro Influencers - Residuals for Likes")
plot_residuals(macro_comments_model, macro_influencers_encoded, 'COMMENTS', predictors, "Macro Influencers - Residuals for Comments")

# Generate comparison table for coefficients and p-values
comparison_table = create_comparison_table(
    [micro_likes_model, micro_comments_model, macro_likes_model, macro_comments_model],
    ["Micro Influencers - Likes", "Micro Influencers - Comments", "Macro Influencers - Likes", "Macro Influencers - Comments"]
)

# Display comparison table
from IPython.display import display
display("Comparison Table for Regression Models", comparison_table)
```
